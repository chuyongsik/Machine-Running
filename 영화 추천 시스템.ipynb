```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title] # get the index of the movie that matches the title
    sim_scores = list(enumerate(cosine_sim[idx])) # sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True) # sort the movies based on the similarity scores lamdda x: x[1] means sort by the second element in the tuple
    sim_scores = sim_scores[1:11] # get the scores of the 10 most similar movies
    movie_indices = [i[0] for i in sim_scores] # get the movie indices
    
    return df_m['title'].iloc[movie_indices] # return the top 10 most similar movies
    
    
    
```

```python
def get_list(x):
    if isinstance(x, list):
        names=[i['name'] for i in x]
        if len(names) >3: # to get the first 3 elements of the list
            names=names[:3]
            return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df_m[feature] = df_m[feature].apply(get_list)
```

```python
df_m[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df_m[feature] = df_m[feature].apply(clean_data)
```

```python
df_m[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres']) # join the keywords, cast, director, and genres into one string
df_m['soup'] = df_m.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df_m['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix,count_matrix)
cosine_sim2
```
